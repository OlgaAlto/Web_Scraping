```python
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib
import pandas as pd
import requests
```

```python
driver = webdriver.Chrome(executable_path='/Users/olga/Downloads/chromedriver')
```

```python
driver.get("https://www.gigantti.fi/INTERSHOP/web/WFS/store-gigantti-Site/fi_FI/-/EUR/CC_RecommendedProductsViewAll-Start?PageletUUID=WAqsGQV9EscAAAFmy6WN8A_L")
content = driver.page_source
```

```python
products=[] #List to store name of the product
prices=[] #List to store price of the product
energy_class=[] #List to store energy class of the product
```

```python
#Getting products description
soup2 = BeautifulSoup(content, 'html.parser')
nameList = soup2.findAll('a', attrs={'class':'product-name'})
for name in nameList:
    products.append(name.text)
print(products)

df1= pd.DataFrame({'Product Name':products}) 
```

```python
df1
```

```python
df1['Product Name'] = df1['Product Name'].str[1:]
df1
```

```python
#Getting products price
soup3 = BeautifulSoup(content, 'html.parser')
nameList2 = soup3.findAll('div', attrs={'class':'product-price'})
for price in nameList2:
    prices.append(price.text)
print(prices)
```

```python
df2= pd.DataFrame({'Price':prices}) 
df2
```

```python
df2['Price'] = df2['Price'].str.replace('\n','')
df2
```

```python
df= pd.concat([df1,df2], ignore_index=True, axis=1)
df
```

```python
df=df.rename(columns={0: "Product", 1: "Price"})
df
```

```python
#Getting products energy class
soup4 = BeautifulSoup(content, 'html.parser')
nameList3 = soup4.findAll('text',attrs={'class':"basic-text-svg"})[0::2]
for ii in nameList3:
    energy_class.append(ii.text)
print(energy_class)
```

```python
df3= pd.DataFrame({'Energy class':energy_class}) 
df3
```

```python
df3['Energy class'] = df3['Energy class'].str.replace('\n','')
df3
```

```python
df= pd.concat([df,df3], ignore_index=True, axis=1)
df
```

```python
df=df.rename(columns={0: "Product", 1: "Price", 2: "Energy class"})
df
```
